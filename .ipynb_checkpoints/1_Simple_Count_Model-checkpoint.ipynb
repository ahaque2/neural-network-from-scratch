### Loading training data

```python
words = open('data/names.txt', 'r').read().splitlines()
```

```python
len(words), words[0:10]
```

### Create Bigrams
#### This is also the training phase for this model 
- The model learns the counts of each pair and infers based on that knowledge)

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
import pandas as pd
bigrams_df = pd.DataFrame(b.keys(), columns = ['ch1', 'ch2'])
bigrams_df['counts'] = b.values()
bigrams_df.shape
```

## Optimizing the representation of bigram pair counts

```python
import torch
```

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

### Encoding and Decoding

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0       ### Instead of <S> and <E>, we use one character for both '.'
itos = {i:s for s,i in stoi.items()}
```

#### Training (using counts)

```python
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    N[ix1, ix2] += 1
```

```python
N.shape, N[0]
```

```python
import matplotlib.pyplot as plt
# %matplotlib inline

def plot_matrix(N):

    plt.figure(figsize=(20,20))
    plt.imshow(N, cmap='BuPu')
    for i in range(27):
        for j in range(27):
            chstr = itos[i] + itos[j]
            plt.text(j, i, chstr, ha="center", va="bottom", color='black')
            plt.text(j, i, N[i, j].item(), ha="center", va="top", color='grey')
    plt.axis('off');
```

### Generating new names 
#### Choosing the character with highest probability to generate next character based on the count matrix N

```python
import sys
names = []
P = (N).float()
for _ in range(10):
    out = []
    ix = 0            #torch.randint(0, 27, (1,)).item()
    while True:
        n = P[ix]    ## we are interested in rows and not columns. Why?
        ix = torch.argmax(n).item()
        # ix = torch.multinomial(n, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        # print("Out ", out)
        if ix == 0:
          break
    print(''.join(out))
    names.append(''.join(out))
# pd.Series([n[-2] for n in names]).value_counts()
```

#### Understanding Multinomial Sampling

```python
g = torch.Generator().manual_seed(42)
p_ = torch.rand(3, generator=g)
# p_ = p_ / p_.sum()
print(p_)

# torch.multinomial(p_, num_samples=100, replacement=True, generator=g)   #.unique(return_counts=True)
```

```python
### Transform the counts to something which can be compared more reliably
### We normalize the count distribution for each character to get a probability distribution
### Each row in P sums up to 1 (hence we have a scale to compare against when sampling)

P = (N).float()      ## Model smoothing: (N+1) instead of (N) [Avoids 0 probability for bigrams that dont exist]
P = P/P.sum(1, keepdims=True)    ## Its important to have keepdims = True to ensure correct broadcasting
P.shape
```

```python
# torch.round(P, decimals = 1)

def plot_matrix(N):

    plt.figure(figsize=(20,20))
    plt.imshow(N, cmap='BuPu')
    for i in range(27):
        for j in range(27):
            chstr = itos[i] + itos[j]
            plt.text(j, i, chstr, ha="center", va="bottom", color='black')
            # plt.text(j, i, N[i, j].item(), ha="center", va="top", color='grey')
    plt.axis('off');
    
import numpy as np
plot_matrix(np.array(P).round(2))
```

```python
import sys
names = []
for _ in range(10):
    out = []
    ix = 0            #torch.randint(0, 27, (1,)).item()
    while True:
        n = P[ix]    ## we are interested in rows and not columns. Why?
        ix = torch.multinomial(n, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        # print("Out ", out)
        if ix == 0:
          break
    print(''.join(out))
    names.append(''.join(out))
# pd.Series([n[-2] for n in names]).value_counts()
```

```python
loss_prob = []
import math
# for t in test_set:    
chs = ['.'] + list(test_set[0]) + ['.']
# chs = ['.'] + list(t) + ['.']
print(chs)
loss = []

for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]

    prob = P[ix1, ix2]         ### Probabity of next token given the first
    loss.append(prob)
    print(f'{ch1}{ch2}: {prob:.4f}')
joint_prob = math.prod(loss)
print(joint_prob)
```

### Computing training loss (negative log likelihood)

```python
loss_logprob = []
# for t in test_set:
chs = ['.'] + list([x for x in words[0]]) + ['.']
# chs = ['.'] + list(t) + ['.']
print(chs)
loss = []

for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]

    prob = P[ix1, ix2]         ### Probabity of next token given the first
    logprob = torch.log(prob)  ## Converting to log (otherwise the product will be very small since all numbers are <1). And backpropagating through log of a product is easier (just the sum of individual logs) 
    loss.append(logprob)
    print(f'{ch1}{ch2}: {prob:.4f}: {logprob:.4f}')
loss_logprob.append((sum(loss)/len(loss)).item())   
sum(loss)/len(loss), sum(loss)
```

### Evaluating training loss over the entire training dataset

```python
log_likelihood = 0.0
n = 0

for w in words:
# for w in ["amanulx"]:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]         ### Probabity of next token given the first
    
    # print(prob)
    # sys.exit()
    logprob = torch.log(prob)  ## Converting to log (otherwise the product will be very small since all numbers are <1). And backpropagating through log of a product is easier (just the sum of individual logs) 
    log_likelihood += logprob  ## Product of all probabilities (we take the sum because we are using logs). 
    n += 1
    # print(f'{ch1}{ch2}: {prob:.4f}: {logprob:.4f}')

print(f'{log_likelihood=}')
nll = -log_likelihood      # Invert the loss to ensure 'lower is better' (since log of < 1 os already negative)
print(f'{nll=}')
print(f'{nll/n}')      ## Normalized log likelihood (loss)
print(n)

# GOAL: maximize likelihood of the data w.r.t. model parameters (statistical modeling)
# equivalent to maximizing the log likelihood (because log is monotonic)
# equivalent to minimizing the negative log likelihood
# equivalent to minimizing the average negative log likelihood

# log(a*b*c) = log(a) + log(b) + log(c)

### Any bigram with probablity > 1/27 means that the model has learnt something 
```
